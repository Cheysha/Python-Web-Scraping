```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import re
```

```python
# load ./Data/review_dataframes.csv
df = pd.read_csv('./Data/review_dataframes.csv')
```

```python
# plat the distribution of review scores
sns.set_style('whitegrid')
sns.countplot(x='Quality', data=df, palette='RdBu')
# print number of 5.0 quality reviews
print('Number of 5.0 quality reviews: ', len(df[df['Quality'] == 5.0]), '\\', len(df) , ' = ', len(df[df['Quality'] == 5.0])/len(df))
# print number of 1.0 quality reviews
print('Number of 1.0 quality reviews: ', len(df[df['Quality'] == 1.0]), '\\', len(df) , ' = ', len(df[df['Quality'] == 1.0])/len(df))
```

```python
# plot the distribution of difficulty scores
sns.set_style('whitegrid')
sns.countplot(x='Difficulty', data=df, palette='RdBu')
# print number of 5.0 difficulty reviews
print('Number of 5.0 difficulty reviews: ', len(df[df['Difficulty'] == 5.0]))
# print number of 1.0 difficulty reviews
print('Number of 1.0 difficulty reviews: ', len(df[df['Difficulty'] == 1.0]))
```

```python
# plot the correlation between difficulty and quality
sns.set_style('whitegrid')
#sns.countplot(x='Difficulty', hue='Quality', data=df, palette='RdBu_r')
sns.countplot(x='Difficulty', hue='Quality', data=df, palette='RdBu')
# count how many 3.0 difficulty reviews are 5.0 quality
print('Number of 3.0 difficulty reviews that are 5.0 quality: ', len(df[(df['Difficulty'] == 3.0) & (df['Quality'] == 5.0)]))
```
